# Question 1

In this case the question "Could you give a definition of the supervisor/advisor?" is analized.

First all responses are in one variable to have only one set.

```python
def convert_list_to_string(org_list, seperator=' '):
    """ Convert list to string, by joining all item in list with given separator.
        Returns the concatenated string """
    return seperator.join(org_list)
```

```python
supervisor_definition = convert_list_to_string(df['Could you give a definition of the supervisor/advisor?'].values.tolist())
supervisor_definitio = supervisor_definition.replace("-a","a")
supervisor_definitio = supervisor_definition.replace("-an","an")
supervisor_definitio = supervisor_definition.replace("-emotional","emotional")
supervisor_definitio = supervisor_definition.replace("-offer","offer")
supervisor_definitio = supervisor_definition.replace("-academic","academic")
supervisor_definition
```

```python
supervisor_definition.split()
```

```python
supervisor_definition_words = nltk.word_tokenize(supervisor_definition)
supervisor_definition_words
```

### Bag of words

The simplest model for analyzing text is just to think about text as an unordered collection of words (bag-of-words). This can generally allow to infer from the text the category, the topic or the sentiment.

From the bag-of-words model we can build features to be used by a classifier, here we assume that each word is a feature that can either be True or False. We implement this in Python as a dictionary where for each word in a sentence we associate True, if a word is missing, that would be the same as assigning False.

```python
{word:True for word in supervisor_definition_words}
```

Using the Python string.punctuation list and the English stopwords we can build better features by filtering out those words

```python
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation)
```

```python
def build_bag_of_words_features_filtered(words):
    return {
        word:1 for word in words \
        if not word in useless_words}
```

```python
bag_of_words = build_bag_of_words_features_filtered(supervisor_definition_words)
```

```python
filtered_words = [word for word in supervisor_definition_words if not word in useless_words]
type(filtered_words)
```

The collection package of the standard library contains a Counter class that is handy for counting frequencies of words in our list:

```python
from collections import Counter

word_counter = Counter(filtered_words)
```

It also has a `most_common()` method to access the words with the higher count:

```python
most_common_words = word_counter.most_common()[:]
most_common_words
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

The word counts are sorted and ploted on Logarithmic axes to check the shape of the distribution. A flatter distribution indicates a large vocabulary while a peaked distribution a restricted vocabulary often due to a focused topic or specialized language.

```python
sorted_word_counts = sorted(list(word_counter.values()), reverse=True)

plt.loglog(sorted_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank");
```

The next plot is the histogram of `sorted_word_counts`, which displays how many words have a count in a specific range.

The distribution is highly peaked at low counts, i.e. most of the words appear which a low count, so we better display it on semilogarithmic axes to inspect the tail of the distribution.

```python
plt.hist(sorted_word_counts, bins=50, log=True);
```
